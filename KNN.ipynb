## Dystans euklidesowy

\begin{align*}
\sqrt {\sum_{i=1}^{n} (q_{i} - p_{i})^{2}} 
\end{align*}

```python
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from sklearn import datasets
from collections import Counter
import numpy as np


iris = datasets.load_iris()
x = iris.data[:, :2]             
y = (iris.target != 0) * 1 
```

```python
plt.scatter(x[:,0],x[:,1], c = y)
```

```python
X_train, X_test, y_train,y_test = train_test_split(x,y,test_size=0.8)      #Ze względu na "proste" dane, test_set to az 80%
```

```python
k = 3
```

\begin{align*}
\sqrt {\sum_{i=1}^{n} (q_{i} - p_{i})^{2}} 
\end{align*}

```python
def dystans_euklidesowy(q,p):
    return np.sqrt(np.sum((p-q)**2))
```

```python
def predykcja(x_test):
    dystans = [dystans_euklidesowy(x_test, x_train) for x_train in X_train]
    indeksy_sasaidow = np.argsort(dystans)[:k]                #Posortowane od najblizszego
    labele_sasiadow = [y_train[i] for i in indeksy_sasaidow]  #Wybieramy indeksy dla k najbliższych 

    najczestsze_labele = Counter(labele_sasiadow).most_common(1)     #Wybieramy ten, którego jest najwięcej w okolicy
    return najczestsze_labele[0][0]
```

```python
y_pred = [predykcja(x) for x in X_test]
```

### Dane treningowe

```python
plt.scatter(X_test[:,0],X_test[:,1], c=y_test)
```

```python
plt.scatter(X_test[:,0],X_test[:,1], c=y_pred)
```
